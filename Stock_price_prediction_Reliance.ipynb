```python
# Import necessary Libraries
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.neural_network import MLPRegressor
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

print("="*70)
print("STOCK PRICE PREDICTION - COMPARATIVE ANALYSIS")
print("Models: MLP | LSTM | GRU")
print("="*70)

# Import the data and remove rows containing NAN values
df = pd.read_csv('Reliance_Stock.csv')
df = df.dropna()
print(f"\nDataset loaded: {len(df)} rows")
print(df.head())

df_close = df['Close']

# Plot original data
plt.figure(figsize=(12, 5))
plt.plot(df_close)
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.title('Reliance Stock Closing Price')
plt.grid(True, alpha=0.3)
plt.savefig('01_Original_Stock_Data.jpg', dpi=300, bbox_inches='tight')
print("✓ Saved: 01_Original_Stock_Data.jpg")
plt.show()

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
df_close = scaler.fit_transform(np.array(df_close).reshape(-1, 1))

# Split the data into train and test split
training_size = int(len(df_close) * 0.75)
test_size = len(df_close) - training_size
train_data, test_data = df_close[0:training_size, :], df_close[training_size:len(df_close), :1]

print(f"\nTraining samples: {training_size}")
print(f"Testing samples: {test_size}")

# Create dataset function
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i + time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)

time_step = 100
x_train, y_train = create_dataset(train_data, time_step)
x_test, y_test = create_dataset(test_data, time_step)

print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")

# Store results
results = {}

# ============================================================================
# MODEL 1: MLP (Multi-Layer Perceptron)
# ============================================================================
print("\n" + "="*70)
print("MODEL 1: TRAINING MLP (Multi-Layer Perceptron)")
print("="*70)

mlp_model = MLPRegressor(
    hidden_layer_sizes=(100, 50, 25),
    activation='relu',
    solver='adam',
    alpha=0.0001,
    batch_size='auto',
    learning_rate='adaptive',
    learning_rate_init=0.001,
    max_iter=500,
    random_state=42,
    early_stopping=True,
    validation_fraction=0.1,
    n_iter_no_change=20,
    verbose=False
)

mlp_model.fit(x_train, y_train)
print(f"MLP Training completed! Iterations: {mlp_model.n_iter_}, Loss: {mlp_model.loss_:.6f}")

# MLP Predictions
mlp_train_predict = mlp_model.predict(x_train).reshape(-1, 1)
mlp_test_predict = mlp_model.predict(x_test).reshape(-1, 1)

# Transform back to original form
mlp_train_predict = scaler.inverse_transform(mlp_train_predict)
mlp_test_predict = scaler.inverse_transform(mlp_test_predict)
y_train_orig = scaler.inverse_transform(y_train.reshape(-1, 1))
y_test_orig = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate metrics
results['MLP'] = {
    'train_rmse': math.sqrt(mean_squared_error(y_train_orig, mlp_train_predict)),
    'test_rmse': math.sqrt(mean_squared_error(y_test_orig, mlp_test_predict)),
    'train_mae': mean_absolute_error(y_train_orig, mlp_train_predict),
    'test_mae': mean_absolute_error(y_test_orig, mlp_test_predict),
    'train_r2': r2_score(y_train_orig, mlp_train_predict),
    'test_r2': r2_score(y_test_orig, mlp_test_predict),
    'train_predict': mlp_train_predict,
    'test_predict': mlp_test_predict
}

# ============================================================================
# MODEL 2: LSTM (Long Short-Term Memory)
# ============================================================================
print("\n" + "="*70)
print("MODEL 2: TRAINING LSTM (Long Short-Term Memory)")
print("="*70)

# Reshape for LSTM
x_train_lstm = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test_lstm = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)

lstm_model = Sequential()
lstm_model.add(LSTM(50, return_sequences=True, input_shape=(100, 1)))
lstm_model.add(LSTM(50, return_sequences=True))
lstm_model.add(LSTM(50))
lstm_model.add(Dense(1))
lstm_model.compile(loss='mean_squared_error', optimizer='adam')

print(lstm_model.summary())

lstm_model.fit(x_train_lstm, y_train, validation_data=(x_test_lstm, y_test),
               epochs=100, batch_size=64, verbose=0)
print("LSTM Training completed!")

# LSTM Predictions
lstm_train_predict = lstm_model.predict(x_train_lstm, verbose=0)
lstm_test_predict = lstm_model.predict(x_test_lstm, verbose=0)

# Transform back to original form
lstm_train_predict = scaler.inverse_transform(lstm_train_predict)
lstm_test_predict = scaler.inverse_transform(lstm_test_predict)

# Calculate metrics
results['LSTM'] = {
    'train_rmse': math.sqrt(mean_squared_error(y_train_orig, lstm_train_predict)),
    'test_rmse': math.sqrt(mean_squared_error(y_test_orig, lstm_test_predict)),
    'train_mae': mean_absolute_error(y_train_orig, lstm_train_predict),
    'test_mae': mean_absolute_error(y_test_orig, lstm_test_predict),
    'train_r2': r2_score(y_train_orig, lstm_train_predict),
    'test_r2': r2_score(y_test_orig, lstm_test_predict),
    'train_predict': lstm_train_predict,
    'test_predict': lstm_test_predict
}

# ============================================================================
# MODEL 3: GRU (Gated Recurrent Unit)
# ============================================================================
print("\n" + "="*70)
print("MODEL 3: TRAINING GRU (Gated Recurrent Unit)")
print("="*70)

# Reshape for GRU (same as LSTM)
x_train_gru = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test_gru = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)

gru_model = Sequential()
gru_model.add(GRU(50, return_sequences=True, input_shape=(100, 1)))
gru_model.add(GRU(50, return_sequences=True))
gru_model.add(GRU(50))
gru_model.add(Dense(1))
gru_model.compile(loss='mean_squared_error', optimizer='adam')

print(gru_model.summary())

early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=0)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001, verbose=0)

gru_model.fit(x_train_gru, y_train, validation_data=(x_test_gru, y_test),
              epochs=100, batch_size=64, verbose=0, callbacks=[early_stop, reduce_lr])
print("GRU Training completed!")

# GRU Predictions
gru_train_predict = gru_model.predict(x_train_gru, verbose=0)
gru_test_predict = gru_model.predict(x_test_gru, verbose=0)

# Transform back to original form
gru_train_predict = scaler.inverse_transform(gru_train_predict)
gru_test_predict = scaler.inverse_transform(gru_test_predict)

# Calculate metrics
results['GRU'] = {
    'train_rmse': math.sqrt(mean_squared_error(y_train_orig, gru_train_predict)),
    'test_rmse': math.sqrt(mean_squared_error(y_test_orig, gru_test_predict)),
    'train_mae': mean_absolute_error(y_train_orig, gru_train_predict),
    'test_mae': mean_absolute_error(y_test_orig, gru_test_predict),
    'train_r2': r2_score(y_train_orig, gru_train_predict),
    'test_r2': r2_score(y_test_orig, gru_test_predict),
    'train_predict': gru_train_predict,
    'test_predict': gru_test_predict
}

# ============================================================================
# RESULTS COMPARISON - TABULAR FORM
# ============================================================================
print("\n" + "="*70)
print("PERFORMANCE COMPARISON - TEST SET")
print("="*70)

# Create comparison dataframe
comparison_df = pd.DataFrame({
    'Model': ['MLP', 'LSTM', 'GRU'],
    'RMSE': [results['MLP']['test_rmse'], results['LSTM']['test_rmse'], results['GRU']['test_rmse']],
    'MAE': [results['MLP']['test_mae'], results['LSTM']['test_mae'], results['GRU']['test_mae']],
    'R² Score': [results['MLP']['test_r2'], results['LSTM']['test_r2'], results['GRU']['test_r2']],
    'R² (%)': [results['MLP']['test_r2']*100, results['LSTM']['test_r2']*100, results['GRU']['test_r2']*100]
})

print(comparison_df.to_string(index=False))
print("="*70)

# Save comparison to CSV
comparison_df.to_csv('Model_Comparison_Results.csv', index=False)
print("\n✓ Saved: Model_Comparison_Results.csv")

# Create detailed comparison with train and test metrics
detailed_comparison = pd.DataFrame({
    'Model': ['MLP', 'LSTM', 'GRU'],
    'Train_RMSE': [results['MLP']['train_rmse'], results['LSTM']['train_rmse'], results['GRU']['train_rmse']],
    'Test_RMSE': [results['MLP']['test_rmse'], results['LSTM']['test_rmse'], results['GRU']['test_rmse']],
    'Train_MAE': [results['MLP']['train_mae'], results['LSTM']['train_mae'], results['GRU']['train_mae']],
    'Test_MAE': [results['MLP']['test_mae'], results['LSTM']['test_mae'], results['GRU']['test_mae']],
    'Train_R²': [results['MLP']['train_r2'], results['LSTM']['train_r2'], results['GRU']['train_r2']],
    'Test_R²': [results['MLP']['test_r2'], results['LSTM']['test_r2'], results['GRU']['test_r2']],
    'Train_R²_%': [results['MLP']['train_r2']*100, results['LSTM']['train_r2']*100, results['GRU']['train_r2']*100],
    'Test_R²_%': [results['MLP']['test_r2']*100, results['LSTM']['test_r2']*100, results['GRU']['test_r2']*100]
})

detailed_comparison.to_csv('Detailed_Model_Comparison.csv', index=False)
print("✓ Saved: Detailed_Model_Comparison.csv")

# Find best model
best_r2_model = comparison_df.loc[comparison_df['R² Score'].idxmax(), 'Model']
best_rmse_model = comparison_df.loc[comparison_df['RMSE'].idxmin(), 'Model']
print(f"\n Best R² Score: {best_r2_model}")
print(f" Best RMSE: {best_rmse_model}")

# ============================================================================
# VISUALIZATION 1: BAR CHART COMPARISON
# ============================================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

models = ['MLP', 'LSTM', 'GRU']
colors = ['#3498db', '#e74c3c', '#2ecc71']

# RMSE Comparison
axes[0].bar(models, [results[m]['test_rmse'] for m in models], color=colors, alpha=0.7, edgecolor='black')
axes[0].set_ylabel('RMSE', fontsize=12, fontweight='bold')
axes[0].set_title('Root Mean Squared Error', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3, axis='y')
for i, v in enumerate([results[m]['test_rmse'] for m in models]):
    axes[0].text(i, v + 0.5, f'{v:.2f}', ha='center', fontweight='bold')

# MAE Comparison
axes[1].bar(models, [results[m]['test_mae'] for m in models], color=colors, alpha=0.7, edgecolor='black')
axes[1].set_ylabel('MAE', fontsize=12, fontweight='bold')
axes[1].set_title('Mean Absolute Error', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3, axis='y')
for i, v in enumerate([results[m]['test_mae'] for m in models]):
    axes[1].text(i, v + 0.5, f'{v:.2f}', ha='center', fontweight='bold')

# R² Score Comparison
axes[2].bar(models, [results[m]['test_r2']*100 for m in models], color=colors, alpha=0.7, edgecolor='black')
axes[2].set_ylabel('R² Score (%)', fontsize=12, fontweight='bold')
axes[2].set_title('R² Score Percentage', fontsize=14, fontweight='bold')
axes[2].grid(True, alpha=0.3, axis='y')
axes[2].set_ylim([0, 100])
for i, v in enumerate([results[m]['test_r2']*100 for m in models]):
    axes[2].text(i, v + 1, f'{v:.2f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('02_Bar_Chart_Comparison.jpg', dpi=300, bbox_inches='tight')
print("\n✓ Saved: 02_Bar_Chart_Comparison.jpg")
plt.show()

# ============================================================================
# VISUALIZATION 2: ACTUAL VS PREDICTED - ALL MODELS
# ============================================================================
plt.figure(figsize=(18, 12))

# Full dataset visualization
for idx, model_name in enumerate(['MLP', 'LSTM', 'GRU'], 1):
    plt.subplot(3, 1, idx)

    # Prepare plot data
    look_back = 100
    trainPredictPlot = np.empty_like(df_close)
    trainPredictPlot[:, :] = np.nan
    trainPredictPlot[look_back:len(results[model_name]['train_predict']) + look_back, :] = results[model_name]['train_predict']

    testPredictPlot = np.empty_like(df_close)
    testPredictPlot[:, :] = np.nan
    testPredictPlot[len(results[model_name]['train_predict']) + (look_back * 2) + 1:len(df_close) - 1, :] = results[model_name]['test_predict']

    # Plot
    plt.plot(scaler.inverse_transform(df_close), label='Actual Data', linewidth=1.5, color='blue', alpha=0.7)
    plt.plot(trainPredictPlot, label='Train Predictions', linewidth=1.5, alpha=0.7, color='green')
    plt.plot(testPredictPlot, label='Test Predictions', linewidth=1.5, alpha=0.7, color='red')
    plt.xlabel('Time', fontsize=11)
    plt.ylabel('Stock Price', fontsize=11)
    plt.title(f'{model_name} Model: Actual vs Predicted (R² = {results[model_name]["test_r2"]:.4f})',
              fontsize=13, fontweight='bold')
    plt.legend(fontsize=10)
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('03_Full_Timeline_All_Models.jpg', dpi=300, bbox_inches='tight')
print("✓ Saved: 03_Full_Timeline_All_Models.jpg")
plt.show()

# ============================================================================
# VISUALIZATION 3: TEST SET COMPARISON - ZOOMED VIEW
# ============================================================================
plt.figure(figsize=(16, 10))

# Individual model plots
for idx, model_name in enumerate(['MLP', 'LSTM', 'GRU'], 1):
    plt.subplot(2, 2, idx)
    plt.plot(y_test_orig, label='Actual', linewidth=2, marker='o', markersize=3, alpha=0.7)
    plt.plot(results[model_name]['test_predict'], label=f'{model_name} Predicted',
             linewidth=2, marker='s', markersize=3, alpha=0.7)
    plt.xlabel('Test Samples', fontsize=11)
    plt.ylabel('Stock Price', fontsize=11)
    plt.title(f'{model_name} - Test Set Predictions\nR²={results[model_name]["test_r2"]:.4f}, RMSE={results[model_name]["test_rmse"]:.2f}',
              fontsize=12, fontweight='bold')
    plt.legend(fontsize=10)
    plt.grid(True, alpha=0.3)

# Combined comparison plot
plt.subplot(2, 2, 4)
plt.plot(y_test_orig, label='Actual', linewidth=2.5, color='black', marker='o', markersize=4, alpha=0.8)
plt.plot(results['MLP']['test_predict'], label='MLP', linewidth=2, alpha=0.7, linestyle='--')
plt.plot(results['LSTM']['test_predict'], label='LSTM', linewidth=2, alpha=0.7, linestyle='--')
plt.plot(results['GRU']['test_predict'], label='GRU', linewidth=2, alpha=0.7, linestyle='--')
plt.xlabel('Test Samples', fontsize=11)
plt.ylabel('Stock Price', fontsize=11)
plt.title('All Models - Test Set Comparison', fontsize=12, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('04_Test_Set_Comparison.jpg', dpi=300, bbox_inches='tight')
print("✓ Saved: 04_Test_Set_Comparison.jpg")
plt.show()

# ============================================================================
# VISUALIZATION 4: SCATTER PLOTS - ACTUAL VS PREDICTED
# ============================================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, model_name in enumerate(['MLP', 'LSTM', 'GRU']):
    axes[idx].scatter(y_test_orig, results[model_name]['test_predict'],
                     alpha=0.6, s=50, color=colors[idx], edgecolors='black', linewidth=0.5)

    # Perfect prediction line
    min_val = min(y_test_orig.min(), results[model_name]['test_predict'].min())
    max_val = max(y_test_orig.max(), results[model_name]['test_predict'].max())
    axes[idx].plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2.5, label='Perfect Prediction')

    axes[idx].set_xlabel('Actual Values', fontsize=12, fontweight='bold')
    axes[idx].set_ylabel('Predicted Values', fontsize=12, fontweight='bold')
    axes[idx].set_title(f'{model_name} - Actual vs Predicted\nR² = {results[model_name]["test_r2"]:.4f}',
                       fontsize=13, fontweight='bold')
    axes[idx].legend(fontsize=10)
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('05_Scatter_Plots_Actual_vs_Predicted.jpg', dpi=300, bbox_inches='tight')
print("✓ Saved: 05_Scatter_Plots_Actual_vs_Predicted.jpg")
plt.show()

# ============================================================================
# VISUALIZATION 5: INDIVIDUAL MODEL PERFORMANCE
# ============================================================================
for model_name in ['MLP', 'LSTM', 'GRU']:
    plt.figure(figsize=(14, 6))

    look_back = 100
    trainPredictPlot = np.empty_like(df_close)
    trainPredictPlot[:, :] = np.nan
    trainPredictPlot[look_back:len(results[model_name]['train_predict']) + look_back, :] = results[model_name]['train_predict']

    testPredictPlot = np.empty_like(df_close)
    testPredictPlot[:, :] = np.nan
    testPredictPlot[len(results[model_name]['train_predict']) + (look_back * 2) + 1:len(df_close) - 1, :] = results[model_name]['test_predict']

    plt.plot(scaler.inverse_transform(df_close), label='Actual Data', linewidth=2, color='blue')
    plt.plot(trainPredictPlot, label='Train Predictions', linewidth=2, alpha=0.7, color='green')
    plt.plot(testPredictPlot, label='Test Predictions', linewidth=2, alpha=0.7, color='red')
    plt.xlabel('Time', fontsize=12)
    plt.ylabel('Stock Price', fontsize=12)
    plt.title(f'{model_name} Model Performance - Actual vs Predicted\nTest R² = {results[model_name]["test_r2"]:.4f}, Test RMSE = {results[model_name]["test_rmse"]:.2f}',
              fontsize=14, fontweight='bold')
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'06_{model_name}_Individual_Performance.jpg', dpi=300, bbox_inches='tight')
    print(f"✓ Saved: 06_{model_name}_Individual_Performance.jpg")
    plt.show()

# ============================================================================
# VISUALIZATION 6: COMPARISON TABLE AS IMAGE
# ============================================================================
fig, ax = plt.subplots(figsize=(12, 4))
ax.axis('tight')
ax.axis('off')

# Create table data
table_data = []
table_data.append(['Model', 'Train RMSE', 'Test RMSE', 'Train MAE', 'Test MAE', 'Train R²', 'Test R²'])
for model in ['MLP', 'LSTM', 'GRU']:
    table_data.append([
        model,
        f"{results[model]['train_rmse']:.4f}",
        f"{results[model]['test_rmse']:.4f}",
        f"{results[model]['train_mae']:.4f}",
        f"{results[model]['test_mae']:.4f}",
        f"{results[model]['train_r2']:.4f}",
        f"{results[model]['test_r2']:.4f}"
    ])

table = ax.table(cellText=table_data, cellLoc='center', loc='center',
                colWidths=[0.12, 0.15, 0.15, 0.15, 0.15, 0.14, 0.14])
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 2.5)

# Style header row
for i in range(7):
    table[(0, i)].set_facecolor('#3498db')
    table[(0, i)].set_text_props(weight='bold', color='white')

# Style data rows with alternating colors
for i in range(1, 4):
    color = '#ecf0f1' if i % 2 == 0 else 'white'
    for j in range(7):
        table[(i, j)].set_facecolor(color)

plt.title('Model Performance Comparison Table', fontsize=16, fontweight='bold', pad=20)
plt.savefig('07_Comparison_Table.jpg', dpi=300, bbox_inches='tight')
print("✓ Saved: 07_Comparison_Table.jpg")
plt.show()

# ============================================================================
# FINAL SUMMARY
# ============================================================================
print("\n" + "="*70)
print("FINAL SUMMARY")
print("="*70)
print("\nTraining Set Performance:")
for model in ['MLP', 'LSTM', 'GRU']:
    print(f"{model:6s} - RMSE: {results[model]['train_rmse']:7.4f} | MAE: {results[model]['train_mae']:7.4f} | R²: {results[model]['train_r2']:.4f}")

print("\nTest Set Performance:")
for model in ['MLP', 'LSTM', 'GRU']:
    print(f"{model:6s} - RMSE: {results[model]['test_rmse']:7.4f} | MAE: {results[model]['test_mae']:7.4f} | R²: {results[model]['test_r2']:.4f}")

print("\n" + "="*70)
print("KEY INSIGHTS:")
print("="*70)
print("• MLP: Simple feedforward network, fastest training")
print("• LSTM: Excellent for sequential data, captures long-term dependencies")
print("• GRU: Balanced performance, fewer parameters than LSTM")
print(f"\n WINNER: {best_r2_model} with R² Score of {comparison_df['R² Score'].max():.4f}")
print("="*70)

print("\n" + "="*70)
print("FILES SAVED:")
print("="*70)
print("CSV Files:")
print("  1. Model_Comparison_Results.csv")
print("  2. Detailed_Model_Comparison.csv")
print("\nJPEG Files:")
print("  1. 01_Original_Stock_Data.jpg")
print("  2. 02_Bar_Chart_Comparison.jpg")
print("  3. 03_Full_Timeline_All_Models.jpg")
print("  4. 04_Test_Set_Comparison.jpg")
print("  5. 05_Scatter_Plots_Actual_vs_Predicted.jpg")
print("  6. 06_MLP_Individual_Performance.jpg")
print("  7. 06_LSTM_Individual_Performance.jpg")
print("  8. 06_GRU_Individual_Performance.jpg")
print("  9. 07_Comparison_Table.jpg")
print("="*70)
```
